## Generating Vocabulary and Classes for English

```python
lm = WordNetLemmatizer() #for getting words

# lists
ourClasses = []
newWords = []
documentX = []
documentY = []
# Each intent is tokenized into words and the patterns and their associated tags are added to their respective lists.
for intent in ourData_english["intents"]:
    for pattern in intent["patterns"]:
        ourTokens = nltk.word_tokenize(pattern)
        newWords.extend(ourTokens)
        documentX.append(pattern)
        documentY.append(intent["tag"])
    
    
    if intent["tag"] not in ourClasses:# add unexisting tags to their respective classes
        ourClasses.append(intent["tag"])

newWords = [lm.lemmatize(word.lower()) for word in newWords if word not in string.punctuation] # set words to lowercase if not in punctuation
newWords = sorted(set(newWords))# sorting words
ourClasses = sorted(set(ourClasses))# sorting classes
```

## Generating Vocabulary and Classes for Arabic

```python
sm = SnowballStemmer(language='arabic')
ourClasses_a = []
newWords_a = []
documentX_a = []
documentY_a = []
# Each intent is tokenized into words and the patterns and their associated tags are added to their respective lists.
for intent in ourData_arabic["intents"]:
    for pattern in intent["patterns"]:
        ourTokens = nltk.word_tokenize(pattern)
        newWords_a.extend(ourTokens)
        documentX_a.append(pattern)
        documentY_a.append(intent["tag"])
    
    
    if intent["tag"] not in ourClasses_a:# add unexisting tags to their respective classes
        ourClasses_a.append(intent["tag"])

newWords_a = [sm.stem(word.lower()) for word in newWords_a if word not in string.punctuation] # set words to lowercase if not in punctuation
newWords_a = sorted(set(newWords_a))# sorting words
ourClasses_a = sorted(set(ourClasses_a))# sorting classes
```

## Getting Numeric Train Data for English/Arabic 

```python
trainingData = [] # training list array
outEmpty = [0] * len(ourClasses)
# BOW model
for idx, doc in enumerate(documentX):
    bagOfwords = []
    text = lm.lemmatize(doc.lower())
    for word in newWords:
        bagOfwords.append(1) if word in text else bagOfwords.append(0)
    
    outputRow = list(outEmpty)
    outputRow[ourClasses.index(documentY[idx])] = 1
    trainingData.append([bagOfwords, outputRow])

random.shuffle(trainingData)
trainingData = num.array(trainingData, dtype=object)# coverting our data into an array afterv shuffling

x_e = num.array(list(trainingData[:, 0]))# first trainig phase
y_e = num.array(list(trainingData[:, 1]))# second training phase

trainingData = [] # training list array
outEmpty = [0] * len(ourClasses_a)
# BOW model
for idx, doc in enumerate(documentX_a):
    bagOfwords = []
    text = sm.stem(doc.lower())
    for word in newWords_a:
        bagOfwords.append(1) if word in text else bagOfwords.append(0)
    
    outputRow = list(outEmpty)
    outputRow[ourClasses_a.index(documentY_a[idx])] = 1
    trainingData.append([bagOfwords, outputRow])

random.shuffle(trainingData)
trainingData = num.array(trainingData, dtype=object)# coverting our data into an array afterv shuffling

x_a = num.array(list(trainingData[:, 0]))# first trainig phase
y_a = num.array(list(trainingData[:, 1]))# second training phase
```

```python
ourNewModel_e = MLPClassifier(random_state=1, max_iter=200,verbose=True,early_stopping=False,learning_rate_init=0.01).fit(x_e,y_e)
ourNewModel_a = MLPClassifier(random_state=1, max_iter=300,verbose=True,early_stopping=False,learning_rate_init=0.01).fit(x_a,y_a)
```

```python
def ourText_english(text): 
  newtkns = nltk.word_tokenize(text.lower())
  newtkns = [lm.lemmatize(word) for word in newtkns]
  #newtkns = [lm.stem(word) for word in newtkns]
  return newtkns

def ourText_arabic(text): 
  newtkns = nltk.word_tokenize(text)
  newtkns = [sm.stem(word) for word in newtkns]
  return newtkns

def wordBag(text, vocab,choice): 
  if choice==1:
    newtkns = ourText_english(text)
  else:
    newtkns = ourText_arabic(text)
  bagOwords = [0] * len(vocab)
  for w in newtkns: 
    for idx, word in enumerate(vocab):
      if word == w: 
        bagOwords[idx] = 1
  return num.array(bagOwords)

def pred_class(text, vocab, labels,choice): 
  bagOwords = wordBag(text, vocab,choice)
  if choice==1:
    ourResult = ourNewModel_e.predict(num.array([bagOwords]))[0]
  else:
    ourResult = ourNewModel_a.predict(num.array([bagOwords]))[0]
  newThresh = 0.2
  yp = [[idx, res] for idx, res in enumerate(ourResult) if res > newThresh]

  yp.sort(key=lambda x: x[1], reverse=True)
  newList = []
  for r in yp:
    newList.append(labels[r[0]])
  return newList

def getRes(firstlist, fJson): 
  tag = firstlist[0]
  listOfIntents = fJson["intents"]
  for i in listOfIntents: 
    if i["tag"] == tag:
      ourResult = random.choice(i["responses"])
      break
  return ourResult
```
